```python
# shuffled data 
shuffled = data.sample(frac=1)
shuffled.head()
```

```python
# check null values
shuffled.isnull().sum()
```

```python
# replance label name in 0 or 1 formate
shuffled.replace(('Iris-setosa', 'Iris-virginica','Iris-versicolor'), (0, 1, 2 ), inplace=True)
```

```python
shuffled.head()
```

```python
shuffled.drop(shuffled.columns[[0]], axis = 1, inplace = True)
# shuffled.reset_index(drop=True,inplace=True)
shuffled
```

```python
shuffled.to_csv('E:\MLOps portfolio/iris-flower-mlops/data_given/iris-flower-species.csv')
```
